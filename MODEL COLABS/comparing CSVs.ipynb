```python
import pandas as pd

def compare_csv(file1, file2):
    """
    Compares two CSV files and prints the rows that are the same and different.

    Args:
        file1 (str): Path to the first CSV file.
        file2 (str): Path to the second CSV file.
    """
    try:
        df1 = pd.read_csv('s_lr_submission.csv')
        df2 = pd.read_csv('s_lr2_submission.csv')
    except FileNotFoundError as e:
        print(f"Error: {e}")
        return

    # Find rows that are present in both dataframes
    same_rows = df1[df1.apply(tuple, axis=1).isin(df2.apply(tuple, axis=1))]

    # Find rows that are different
    diff_rows = df1[~df1.apply(tuple, axis=1).isin(df2.apply(tuple, axis=1))]

    print("Same Rows:\n", same_rows)
    print("\nDifferent Rows:\n", diff_rows)

if __name__ == '__main__':
    file1_path = 'file1.csv'
    file2_path = 'file2.csv'
    compare_csv(file1_path, file2_path)
```
