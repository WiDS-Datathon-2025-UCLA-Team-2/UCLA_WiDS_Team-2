```python
def drop_uncommon_columns(df1, df2):
    # Find common columns
    common_columns = df1.columns.intersection(df2.columns)

    # Keep only common columns
    df1 = df1[common_columns]
    df2 = df2[common_columns]

    return df1, df2

train_df, test_df = drop_uncommon_columns(train_df, test_df)

test_df
```

##ADHD Model

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score
```

```python
y = adhd
X = train_df.drop(columns=['participant_id'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Calculate scale_pos_weight to handle class imbalance
scale_pos_weight = len(y_train[y_train == 0]) / len(y_train[y_train == 1])
```

```python
model = XGBClassifier(
    objective='binary:logistic',   # Specify the learning task and the corresponding loss function.
    colsample_bytree=0.3,     # Fraction of features used in each tree. Lower values reduce overfitting.
    learning_rate=0.1,        # Step size shrinkage used to prevent overfitting.
    max_depth=5,              # Maximum depth of each tree. Higher values can lead to overfitting.
    alpha=10,                 # L1 regularization term (Lasso). Higher values add more regularization.
    n_estimators=10,           # Number of boosting rounds (trees) in the ensemble.
    scale_pos_weight=scale_pos_weight
)
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred = model.predict(X_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print(f"F1 Score: {f1:.4f}")
```

##Hyperparameter tuning with GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [100, 300, 500, 800],  # Number of trees in the model
    'max_depth': [3, 5, 7, 10],           # Maximum depth of the trees
    'learning_rate': [0.01, 0.05, 0.1, 0.2],  # Step size to prevent overfitting
    'subsample': [0.7, 0.8, 0.9, 1.0],      # Fraction of samples used per tree
    'colsample_bytree': [0.7, 0.8, 1.0],      # Fraction of features used per tree
    'gamma': [0, 0.1, 0.2, 0.3],  # Regularization term to avoid overfitting
    'scale_pos_weight': [1, 2, 3],  # Adjust for class imbalance (especially for binary classification)
    'alpha': [0, 0.1, 0.5, 1],  # L1 regularization
    'lambda': [0, 0.1, 0.5, 1]  # L2 regularization
}
```

```python
# Perform grid search with cross-validation
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, scoring='accuracy', cv=3, verbose=1) # too long to run trying randomized search
```

```python
# Best parameters and score
print("Best Parameters:", grid_search.best_params_)
print("Best Accuracy:", grid_search.best_score_)
```
