```python
import pandas as pd
import numpy as np
train_i = pd.read_excel('TRAIN_CATEGORICAL_METADATA.xlsx')
train_ii = pd.read_excel('TRAIN_QUANTITATIVE_METADATA.xlsx')
train_iii = pd.read_csv('TRAIN_FUNCTIONAL_CONNECTOME_MATRICES.csv')
solutions = pd.read_excel('TRAINING_SOLUTIONS.xlsx')
```

```python
print(sum(np.sum(train_iii.isnull(), axis = 0)))
train_iii
```

```python
print(np.sum(train_ii.isnull(), axis = 0))
```

```python
col ='MRI_Track_Age_at_Scan'
average_col = train_ii[col].mean()
train_ii[col] = train_ii[col].fillna(average_col) # replaced all Nan values with average
train_ii
```

```python
num_null = np.sum(train_i.isnull(), axis = 0)
train_i
```

```python
for col in num_null.index:
  if num_null[col] > 0:
    average_col = train_i[col].mean()
    train_i[col] = train_i[col].fillna(average_col)
```

```python
sum(np.sum(train_i.isnull(), axis = 0))
```

```python
solutions
```

```python
train_i.nunique() #the number of unique values broken down by each column
```

```python
participant = train_i['participant_id']
df = train_i.drop('participant_id', axis=1)
df=df.astype('int')
df
```

```python
train_i_encoded = pd.get_dummies(df, columns=df.columns, dtype='int')
train_i_encoded
```

```python
train_i_encoded.columns
```

```python
#checking for correct one hot encoded
count=0
for val in train_i['Basic_Demos_Enroll_Year']:
  if val == 2016:
    count+=1
count2=0
for val in train_i_encoded['Basic_Demos_Enroll_Year_2016']:
  if val == 1:
    count2+=1

if count==count2:
  print(True)
```

```python
all_cols = []
all_cols.append(train_i_encoded.columns)
all_cols.append(train_ii.columns)
all_cols.append(train_iii.columns)
all_cols
```

```python
participant
```

```python
df_combined = pd.concat([participant, train_i_encoded], axis=1)
df_combined
```

```python
merged_df = df_combined.merge(train_ii, on='participant_id', how='outer').merge(train_iii, on='participant_id', how='outer')
for col in merged_df.columns:
  if merged_df[col].isna().sum() != 0:
    print(merged_df[col])
```

```python
final_data = merged_df.merge(solutions, on='participant_id', how='outer')
final_data
```

```python
part_id = final_data['participant_id']
final_data.drop(columns=['participant_id'], inplace=True)
final_data.astype(float)
```

```python
for col in final_data.columns:
  if final_data[col].isna().sum() != 0:
    print(final_data[col])
```

```python
df_combined = pd.concat([part_id, final_data], axis=1)
df_combined
```

```python
filename = 'train_data.csv'
df_combined.to_csv(filename) #cannot export to excel since there are too many columns
```
