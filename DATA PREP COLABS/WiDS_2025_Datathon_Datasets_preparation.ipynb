##Handling NAs, Categorical Vals in Test DataFrames

```python
# LOAD TEST CATEGORICAL DATA
path_test_categorical = "/content/drive/My Drive/UCLA_WiDS_Team_2/DATA/TEST/TEST_CATEGORICAL.xlsx"
test_categorical = pd.read_excel(path_test_categorical)
test_categorical.head(10)
```

```python
print(test_categorical.nunique())  # Check number of unique values per column
```

```python
print(test_categorical.isna().sum())
```

```python
# Calculate the count of missing values per column
missing_data = test_categorical.isna().sum()

# Calculate the percentage of missing values per column
missing_percentage = (missing_data / len(test_categorical)) * 100

# Print the percentage of missing values for each column
print(missing_percentage)
```

```python
test_categorical.info()
```

Since these are categorical values, that contain 4 categories, I will drop missing rows, since imputing it with the mean would make another category during One-hot-encoding, which is not useful

```python
test_categorical.dropna(inplace = True)
```

```python
test_categorical.isna().sum().sum()
```

```python
participant = test_categorical['participant_id']
df = test_categorical.drop('participant_id', axis=1)
test_categorical = pd.get_dummies(df, columns=df.columns, dtype='int')
test_categorical
```

```python
test_categorical['participant_id'] = participant
test_categorical
```

```python
# LOAD TEST QUANTITATIVE DATA

path_test_quantitative = "/content/drive/My Drive/UCLA_WiDS_Team_2/DATA/TEST/TEST_QUANTITATIVE_METADATA.xlsx"
test_quantitative = pd.read_excel(path_test_quantitative)
test_quantitative.head()
```

```python
print(test_quantitative.isna().sum())
```

```python
test_quantitative.info()
```

For Quantitative data I will impute NAs with mean, instea of dropping them, because having integer inplaced with mean would contain more data

```python
# Preserve the ID column
id_column = test_quantitative['participant_id']

test_qua_no_id = test_quantitative.drop(columns=['participant_id'])

# Impute missing values with the mean for numeric columns
test_qua_no_id.fillna(test_qua_no_id.mean(), inplace=True)

# Recombine the ID column back to the DataFrame
test_quantitative = pd.concat([id_column, test_qua_no_id], axis=1)

print(test_quantitative.isna().sum())
```

```python
test_quantitative.isna().sum().sum()
```

```python
# LOAD TEST Functional Connectome Matrices

path_test_FCM = "/content/drive/My Drive/UCLA_WiDS_Team_2/DATA/TEST/TEST_FUNCTIONAL_CONNECTOME_MATRICES.csv"
test_FCM = pd.read_csv(path_test_FCM)
test_FCM.head()
```

```python
print(test_FCM.isna().sum().sum())
```

Combine dataframes

```python
test_cat_FCM = pd.merge(test_categorical, test_FCM, on = 'participant_id')
```

```python
test_df = pd.merge(test_cat_FCM, test_quantitative, on = 'participant_id')
```
